```python
import pandas as pd
import getAASeq
```

```python
test_path = "C:\\Users\\wolz-\\Downloads\\wetransfer_test_subset-tsv_2025-11-12_1152\\test_subset.tsv"
training_path = "C:\\Users\\wolz-\\Downloads\\wetransfer_test_subset-tsv_2025-11-12_1152\\train_subset.tsv"
validation_path = "C:\\Users\\wolz-\\Downloads\\wetransfer_test_subset-tsv_2025-11-12_1152\\val_subset.tsv"

df_test = pd.read_csv(test_path, sep="\t")
df_training = pd.read_csv(training_path, sep="\t")
df_validation = pd.read_csv(validation_path, sep="\t")
```

```python
df_test = df_test.loc[:, ['pdb_fn', 'mutations', 'total_score']]
df_training = df_training.loc[:, ['pdb_fn', 'mutations', 'total_score']]
df_validation = df_validation.loc[:, ['pdb_fn', 'mutations', 'total_score']]
df_test['set'] = "test"
df_training['set'] = "training"
df_validation['set'] = "validation"
big_df = pd.concat([df_test, df_training, df_validation], ignore_index=True)
```

```python
pdb_map = {}
for pdb_id in big_df["pdb_fn"].unique():
    pdb_map[pdb_id] = getAASeq.get_AA(pdb_id[0:4])
```

```python
big_df["Sequence"] = big_df["pdb_fn"].map(pdb_map)
big_df.head()
```
